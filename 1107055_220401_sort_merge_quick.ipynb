Merge Sort

```python
def merge_sort(arr, l, r):
    if l < r:
      mid = l+(r-l)//2
      merge_sort(arr, l, mid)
      merge_sort(arr, mid+1, r)
      merge(arr, l, mid, r)

def merge(arr, l, mid, r):
    n1 = mid - l + 1
    n2 = r - mid
    left = [0] * (n1)
    right = [0] * (n2)
 
    for i in range(0, n1):
      left[i] = arr[l + i]

    for j in range(0, n2):
      right[j] = arr[mid + 1 + j]
 
    i = 0 
    j = 0
    k = l 
 
    while i < n1 and j < n2:
      if left[i] <= right[j]:
        arr[k] = left[i]
        i += 1
      else:
        arr[k] = right[j]
        j += 1
      k += 1
 
    while i < n1:
      arr[k] = left[i]
      i += 1
      k += 1
    while j < n2:
      arr[k] = right[j]
      j += 1
      k += 1
```

```python
arr = np.loadtxt('/content/couting.txt')

start = datetime.now()
merge_sort(arr, 0, len(arr)-1)
end = datetime.now()

plt.plot(arr)
plt.show()
print('O tempo de execução foi:',end - start)
```

Quick Sort

```python
def partition(arr, l, r):
    i = (l - 1)
    pivot = arr[r]
    for j in range(l, r):
      if arr[j] <= pivot:
        i = i + 1
        arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[r] = arr[r], arr[i + 1]
    return (i+1)
 
def quick_sort(arr, l, r):
    if len(arr) == 1:
      return arr
    if l < r:
      pivot = partition(arr, l, r)
      quick_sort(arr, l, pivot-1)
      quick_sort(arr, pivot+1, r)
```

```python
arr = np.loadtxt('/content/couting.txt')

l = 0
r = len(arr)-1

start = datetime.now()
quick_sort(arr, l, r)
end = datetime.now()

plt.plot(arr)
plt.show()
print('O tempo de execução foi:',end - start)
```
